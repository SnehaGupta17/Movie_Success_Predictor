```python
# Load Datasets
bollywood_df = pd.read_csv("Final Bollywood.csv")
hollywood_df = pd.read_csv("Final Hollywood.csv")
```

```python
# Add Industry Column in each Datasets
bollywood_df['Industry'] = 'Bollywood'
hollywood_df['Industry'] = 'Hollywood'
```

```python
hollywood_df.isnull().sum()
```

```python
hollywood_df.head(310)
```

### Combine into one Dataframe

```python
# Combine into one Dataframe
combined_df = pd.concat([bollywood_df, hollywood_df], ignore_index=True)
```

```python
combined_df
```

```python
combined_df.isnull().sum()
```

```python
# Show rows where Genre is actually missing in Hollywood
missing_genre_rows = hollywood_df[hollywood_df['Genre'].isnull()]
print("Missing Genre rows in Hollywood:")
print(missing_genre_rows)
print("Total:", len(missing_genre_rows))
```

```python
# Fill missing values in Genre with 'Unknown'
combined_df['Genre'].fillna('Unknown', inplace=True)
```

```python
# list to store Hit or Flop
success_status = [] 

for i in range(len(combined_df)):
    revenue = combined_df.loc[i, 'Revenue($)']
    budget = combined_df.loc[i, 'Budget($)']
    
    if revenue >= budget:
        success_status.append('Hit')
    else:
        success_status.append('Flop')

# Add it as a new column
combined_df['Success_Status'] = success_status
```

```python
combined_df['Success_Status'].value_counts()
```

```python
# Clean spaces around each genre name
combined_df['Genre'] = combined_df['Genre'].apply(
    lambda x: ','.join([g.strip().lower() for g in x.split(',')])
)
```

```python
combined_df
```

```python
# Create clear genre list
# Remove spaces and split by comma
combined_df['Genre_List'] = combined_df['Genre'].apply(lambda x: x.replace(" ", "").split(','))
```

```python
combined_df
```

### Encoding

```python
# Multi Label 
from sklearn.preprocessing import MultiLabelBinarizer

# Use MultiLabelBinarizer to one-hot encode the genre list
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(combined_df['Genre_List'])
```

```python
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)
```

```python
genre_counts = genre_df.sum().sort_values(ascending=False)
```

```python
top_20_genres = genre_counts.head(20).index.tolist()
```

```python
genre_top_20_df = genre_df[top_20_genres]
```

```python
combined_df = pd.concat([combined_df, genre_top_20_df], axis=1)
```

```python
# Also drop the original 'Genre' and 'Genre_List' columns (not needed anymore)
combined_df.drop(['Genre', 'Genre_List'], axis=1, inplace=True)
```

```python
combined_df
```

```python
# Success Status and industry
# hit --> 1   Flop-->0
from sklearn.preprocessing import LabelEncoder

# Create the encoder
le = LabelEncoder()

# Apply label encoding on the target column
combined_df['Success_Status'] = le.fit_transform(combined_df['Success_Status'])
```

```python
combined_df
```

```python
combined_df['orig_lang'] = combined_df['orig_lang'].str.strip().str.lower()
```

```python
combined_df['orig_lang'].value_counts().head(10)
```

```python
combined_df['country'] = combined_df['country'].str.strip().str.upper()
```

```python
combined_df['country'].value_counts().head(10)
```

```python
top_langs = combined_df['orig_lang'].value_counts().head(10).index
top_countries = combined_df['country'].value_counts().head(10).index

# Replace all others with 'other'
combined_df['orig_lang'] = combined_df['orig_lang'].apply(lambda x: x if x in top_langs else 'other')
combined_df['country'] = combined_df['country'].apply(lambda x: x if x in top_countries else 'other')

# One-hot encode
combined_df = pd.get_dummies(combined_df, columns=['orig_lang', 'country'])
```

```python
# One-hot encode for industry
combined_df = pd.get_dummies(combined_df, columns=['Industry'])
```

```python
combined_df
```

```python
combined_df = combined_df.replace({True: 1, False: 0})
```

```python
combined_df
```

### Shuffle the dataset

```python
from sklearn.utils import shuffle
combined_df = shuffle(combined_df, random_state=42)
combined_df.reset_index(drop=True, inplace=True)
```

### Scaling : Standardization

```python
# Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale only the continuous numerical columns
numeric_cols = ['Revenue($)', 'Budget($)', 'score']
combined_df[numeric_cols] = scaler.fit_transform(combined_df[numeric_cols])
```

```python
combined_df
```

```python
combined_df['Success_Status'].value_counts()
```

# Model building

```python
from sklearn.model_selection import train_test_split

x = combined_df.drop(['Title', 'Date', 'Success_Status'], axis=1)
y = combined_df['Success_Status']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

### Use SMOTE to balance imbalanced data

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train_resampled, y_train_resampled)
```

```python
# Predict
y_pred = log_reg.predict(x_test)
```

### Accuracy for logistic regression

```python
from sklearn.metrics import classification_report, accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\n",classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```

### Top 20 Features Influencing Movie Success

```python
# Get coefficients and feature names
coefficients = log_reg.coef_[0]
features = x.columns

# Sort by absolute importance
indices = np.argsort(abs(coefficients))[::-1]

# Plot top 20
plt.figure(figsize=(12, 6))
plt.barh(range(20), coefficients[indices[:20]], align='center')
plt.yticks(range(20), [features[i] for i in indices[:20]])
plt.xlabel("Feature Importance (Coefficient Value)")
plt.title("Top 20 Features Influencing Movie Success")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from scipy.stats import uniform

# Define param distributions
param_dist = {
    'C': uniform(0.01, 10),
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l2']  # 'l1' is not supported with lbfgs
}

# Initialize model
model = LogisticRegression(max_iter=1000)

# Use RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=10,               # Try 10 random combinations
    cv=3,                    # Reduce folds to 3 for speed
    scoring='accuracy',
    n_jobs=-1,               # Use all CPU cores
    random_state=42
)

# Fit on training data
random_search.fit(x_train_resampled, y_train_resampled)
```

```python
print("Best Parameters:", random_search.best_params_)
```

```python
best_model = random_search.best_estimator_
y_pred = best_model.predict(x_test)
```

### Acuracy after RandomSearch CV

```python
print("Predicted values:\n", y_pred)

# Accuracy
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
# Classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```
